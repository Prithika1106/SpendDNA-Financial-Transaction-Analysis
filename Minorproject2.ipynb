```python
# AI-assisted: Library import and project setup

import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# AI-assisted: Loading dataset

df = pd.read_csv("Data set for DADS June.csv")

print("Dataset Loaded Successfully!")
print()

print("Shape:", df.shape)
print()

df.head()
```

```python
print("="*60)
print("DATASET INFORMATION")
print("="*60)

print("\nColumns")
print(df.columns.tolist())

print("\nData Types")
print(df.dtypes)

print("\nMissing Values")
print(df.isnull().sum())

print("\nDuplicate Rows")
print(df.duplicated().sum())
```

```python
# AI-assisted: Parsing multiple date formats

df["Date"] = pd.to_datetime(
    df["Date"],
    errors="coerce",
    dayfirst=True
)

print("Unparsed Dates:", df["Date"].isna().sum())
```

```python
# AI-assisted: Cleaning Amount column

df["Amount"] = (
    df["Amount"]
    .astype(str)
    .str.replace("₹", "", regex=False)
    .str.replace("Rs.", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

df["Amount"] = pd.to_numeric(
    df["Amount"],
    errors="coerce"
)

print("Unparsed Amounts:", df["Amount"].isna().sum())
```

```python
# AI-assisted: Standardizing transaction types

df["Type"] = (
    df["Type"]
    .astype(str)
    .str.lower()
    .str.strip()
)

type_mapping = {
    "dr": "debit",
    "debit": "debit",
    "cr": "credit",
    "credit": "credit"
}

df["Type"] = df["Type"].replace(type_mapping)

print(df["Type"].value_counts())
```

```python
# AI-assisted: Cleaning Mode column

df["Mode"] = (
    df["Mode"]
    .replace("", np.nan)
)

print(df["Mode"].isna().sum())
```

```python
duplicates_before = df.duplicated().sum()

df = df.drop_duplicates()

duplicates_after = df.duplicated().sum()

print("Duplicates Removed:", duplicates_before)
print("Remaining Duplicates:", duplicates_after)
```

```python
rows_before = len(df)

df = df.dropna(subset=["Date", "Amount"])

rows_after = len(df)

print("Rows Removed:", rows_before - rows_after)
```

```python
# AI-assisted: Creating date-related features

df["Month"] = df["Date"].dt.month_name()

df["Month_Number"] = df["Date"].dt.month

df["Day"] = df["Date"].dt.day_name()

df["Year"] = df["Date"].dt.year

df["Hour"] = df["Time"].str[:2].astype(int)

df.head()
```

```python
# AI-assisted: Vendor keyword dictionary

vendor_dict = {

    "Swiggy": ["SWIGGY", "BUNDL"],

    "Zomato": ["ZOMATO"],

    "Zepto": ["ZEPTO"],

    "Blinkit": ["BLINKIT"],

    "BigBasket": ["BIGBASKET"],

    "Instamart": ["INSTAMART"],

    "Amazon": ["AMAZON", "AMZN", "AMAZONPAY"],

    "Flipkart": ["FLIPKART"],

    "Myntra": ["MYNTRA"],

    "Ajio": ["AJIO"],

    "Nykaa": ["NYKAA"],

    "Uber": ["UBER"],

    "Ola": ["OLA"],

    "Rapido": ["RAPIDO"],

    "IRCTC": ["IRCTC"],

    "BookMyShow": ["BOOKMYSHOW"],

    "Netflix": ["NETFLIX"],

    "Spotify": ["SPOTIFY"],

    "YouTube": ["YOUTUBE"],

    "Hotstar": ["HOTSTAR"],

    "Prime Video": ["PRIME"],

    "Jio": ["JIO"],

    "Airtel": ["AIRTEL"],

    "BSNL": ["BSNL"],

    "ACT": ["ACT"],

    "Bescom": ["BESCOM"],

    "Water Bill": ["BWSSB"],

    "HP Fuel": ["HPCL", "HP PETROL"],

    "Indian Oil": ["IOCL"],

    "Shell": ["SHELL"],

    "Reliance": ["RELIANCE"],

    "DMart": ["DMART"],

    "More": ["MORE"],

    "Starbucks": ["STARBUCKS"],

    "Cafe Coffee Day": ["CCD"],

    "Third Wave Coffee": ["THIRD WAVE"],

    "Zerodha": ["ZERODHA"],

    "Groww": ["GROWW"],

    "Paytm": ["PAYTM"],

    "PhonePe": ["PHONEPE"],

    "Google Pay": ["GPAY"],

    "Cash Withdrawal": ["ATM-WDL", "ATM"],

    "P2P Transfer": ["OKAXIS", "OKHDFC", "OKSBI", "@OK", "@YBL"]
}
```

```python
# AI-assisted: Vendor extraction

def extract_vendor(description):

    description = str(description).upper()

    for vendor, keywords in vendor_dict.items():

        for keyword in keywords:

            if keyword in description:
                return vendor

    return "Uncategorised"
```

```python
df["Vendor"] = df["Description"].apply(extract_vendor)

print("Vendor column created successfully!")
```

```python
# AI-assisted: Computing category statistics for anomaly detection

category_mean = (
    debit_df
    .groupby("Category")["Amount"]
    .transform("mean")
)

category_mean.head()
```

```python
category_std = (
    debit_df
    .groupby("Category")["Amount"]
    .transform("std")
)

category_std.head()
```

```python
debit_df["Z_Score"] = (
    (debit_df["Amount"] - category_mean)
    / category_std
)

debit_df["Z_Score"] = debit_df["Z_Score"].fillna(0)

debit_df.head()
```
